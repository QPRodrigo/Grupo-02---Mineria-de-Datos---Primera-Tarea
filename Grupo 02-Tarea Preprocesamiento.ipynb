### 06 Buscar nulos y reemplazar

```python

myData = pd.read_csv('Niños LIMA.csv')

# Buscaremos todas las columnas con valor object
tipos = myData.columns.to_series().groupby(myData.dtypes).groups
colText = tipos[np.dtype('object')]
print(colText)
print(len(colText))
```

```python
# Ya que todas poseen el valor object, buscamos valores nulos y los reemplazamos por valores similares
for c in colText:
    mode = myData[c].mode()[0]
    myData[c] = myData[c].fillna(mode)
myData

#Nos mostrará la tabla con los valores reemplazados:
```

### 07 Utilizando operadores como > < = borrar todas las instancias mayores que ..

```python
# Primer modo aplicado para la resolución:

myData = pd.read_csv('Niños LIMA.csv')
myData.loc[0:15,'EdadMeses']
```

```python
filtro = myData['EdadMeses'] < 50
#Con el operador == eliminamos las instancias donde la edad es mayor a50
myData = myData[filtro]
myData.loc[0:15,'EdadMeses']

#Nos muestra las instancias menor o igual a 50:
```

```python
# Segundo modo aplicado para la resolución:

myData = pd.read_csv('Niños LIMA.csv')
myData.loc[0:20,'EdadMeses']
```

```python
for i in range(0, len(myData)):
    try:
        # Elimina los mayores de 50, menores 10 igual de 25
        if i<= 20:   
            if myData.loc[i,'EdadMeses'] > 50 or myData.loc[i,'EdadMeses'] < 10 or myData.loc[i,'EdadMeses'] == 25: 
                myData = myData.drop([i])
    except KeyError as err:
        print(f"Error.. {i}")
myData.loc[0:20,'EdadMeses']

#Nos muestra las instancias menor o igual a 50, mayores que 10 y diferente de 25:
```

### 10 Multiplicadores basado en condiciones

```python
# Primer modo aplicado para la resolución:

myData = pd.read_csv('Niños LIMA.csv')
myData.loc[0:10,["Talla"]]
```

```python
#Se le multiplicara a cada talla (del rango del 0 a 10) que sea mayor a 75 por 0.01
for i in range(0, len(myData)):
    try:
        # 
        if i<= 10:  
            if myData.loc[i,'Talla'] > 75:
                myData.loc[i,"Talla"] = myData.loc[i,"Talla"] * 0.01
    except KeyError as err:
        print(f"Error.. {i}")
myData.loc[0:10,["Talla"]]

#Se muestra las tallas que eran mayores a 75, multiplicadas por 0.01:
```

```python
# Segundo modo aplicado para la resolución:

myData = pd.read_csv('Niños LIMA.csv')
myData.loc[0:10,["Talla"]]
```

```python
#Se le multiplicara a cada talla (del rango del 0 a 10) que sea mayor a 75 por 0.01
myData.loc[myData['Talla'] > 75, "Talla"] = myData['Talla']*0.01
myData.loc[0:10,["Talla"]]
#Se muestra las tallas que eran mayores a 75, multiplicadas por 0.01:
```
